## Reading my dataset and assigning it to variable df

```python
df = spark.read.csv("/home/analytics4220/Kevin S-A/attacks%5B1%5D.csv", header=True)
df
```

## Selected columns I will be using

```python
filtered_df = df[["Country", "Area", "Location", "Time"]]
```

## Calculate the amount of attacks in different countries

```python
filtered_df.groupBy("Country").count().sort('count', ascending = False).show()
```

## Calculate the number of times a certain time appears

```python
filtered_df.groupBy("Time").count().sort('count', ascending = False).show()
```

## Calculate the number of times a certain location appears

```python
filtered_df.groupBy("Location").count().sort('count', ascending = False).show()
```
